1Ô∏è‚É£ Bubble Sort Theory

Bubble Sort is a simple comparison-based sorting algorithm.<br>

It works by repeatedly swapping adjacent elements if they are in the wrong order.<br>

The largest elements ‚Äúbubble‚Äù to the end in each pass, hence the name.<br>

Sorting continues until no swaps are needed.<br>

Steps (Ascending Order):

Start at the first element.

Compare the current element with the next element.

Swap if the current element is greater than the next.

Move to the next element and repeat until the end of the array.

Repeat passes for the remaining unsorted portion of the array until sorted.

Example (Array: [5, 1, 4, 2, 8])

Pass 1:

Compare 5 and 1 ‚Üí swap ‚Üí [1, 5, 4, 2, 8]<br>
Compare 5 and 4 ‚Üí swap ‚Üí [1, 4, 5, 2, 8]<br>
Compare 5 and 2 ‚Üí swap ‚Üí [1, 4, 2, 5, 8]<br>
Compare 5 and 8 ‚Üí no swap ‚Üí [1, 4, 2, 5, 8]<br>


Pass 2:

Compare 1 and 4 ‚Üí no swap <br>
Compare 4 and 2 ‚Üí swap ‚Üí [1, 2, 4, 5, 8]<br>
Compare 4 and 5 ‚Üí no swap<br>
Compare 5 and 8 ‚Üí no swap<br>


Pass 3:

Compare 1 and 2 ‚Üí no swap<br>
Compare 2 and 4 ‚Üí no swap<br>
Compare 4 and 5 ‚Üí no swap<br>
Compare 5 and 8 ‚Üí no swap<br>


‚úÖ Array is now sorted: [1, 2, 4, 5, 8]

```python
# Increasing order of list
def bubble_sort_icr(arr):
    n=len(arr)
    for i in range(n):
        for j in range(n-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
arr1=[90,2,5,12,32]
res=bubble_sort_icr(arr1)
print(res)
```

```python
# Decreasing order of list
def bubble_sort_dcr(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]<arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
arr2=[90,2,5,12,32]
res1=bubble_sort_dcr(arr2)
print(res1)
```

1Ô∏è‚É£ Loops Analysis

Outer loop: for i in range(n) ‚Üí runs n times

Inner loop: for j in range(n-1) ‚Üí runs n-1 times each outer loop iteration

Total comparisons = n * (n-1) ‚âà n¬≤

```python
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        swap=False
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swap=True
        if not swap:
            break
    return arr
arr3=[90,2,5,12,32]  
res3=bubble_sort(arr3)
print(res3)
```

```python
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min_index=i
        for j in range(i+1,n):
            if arr[min_index]>arr[j]:
                min_index=j
        arr[i],arr[min_index]=arr[min_index],arr[i]
    return arr
arr1=[20,1,3,45,11]
result1=selection_sort(arr1)
print(result1)
```

```python
def selection(arr):
    n=len(arr)
    for i in range(n):
        max_index=i
        for j in range(i,n):
            if arr[max_index]<arr[j]:
                max_index=j
        temp=arr[i]
        arr[i]=arr[max_index]
        arr[max_index]=temp
    return arr
arr2=[44,1,3,2,5,6120]
result2=selection(arr2)
print(result2)
```

# Insertion sort

Assume the first element is already sorted.

Take the next element ‚Üí call it key.

Compare key with elements in the sorted part (left side).

Shift all elements that are larger than key one position to the right.

Insert the key in the correct empty spot.

Repeat for every element.

```python
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        key=arr[i]
        j=i-1
        while j>=0 and arr[j]>key:
            arr[j+1]=arr[j]
            j-=1
        arr[j+1]=key
    return arr
arr3=[22,25,28,29,1]
result3=insertion_sort(arr3)
print(result3)
```

1. Best Case Time Complexity ‚Äî O(n)

This happens when the array is already sorted.

Why O(n)?

For each element, the inner while loop checks only one time and stops immediately.

So only n‚àí1 comparisons ‚Üí linear time.

Example:

[1, 2, 3, 4, 5]

‚≠ê 2. Worst Case Time Complexity ‚Äî O(n¬≤)

This happens when the array is reverse sorted.

Example:

[9, 7, 5, 3, 1]


Why O(n¬≤)?

Each new key must be compared with all previous elements.

Number of comparisons:

(n‚àí1) + (n‚àí2) + ... + 2 + 1 = n(n‚àí1)/2


This is proportional to n¬≤ ‚Üí quadratic time.

```python
def partition(arr,low,high):
    pivot=arr[high]
    i=low-1
    for j in range(low,high):
        if arr[j]>=pivot:
            i+=1
            arr[i],arr[j]=arr[j],arr[i]
    arr[i+1],arr[high]=arr[high],arr[i+1]
    return i+1
def quick_sort(arr,low,high):
    if low<high:
        pi=partition(arr,low,high)
        quick_sort(arr,low,pi-1)
        quick_sort(arr,pi+1,high)
arr=[100,1,0,2,5,7]
quick_sort(arr,0,len(arr)-1)
print(arr)
```

üß† 1. What is Merge Sort? (Theory)

Merge Sort is a Divide and Conquer algorithm:

Divide the array into two halves

Conquer ‚Üí Recursively sort both halves

Combine ‚Üí Merge the sorted halves

‚úî Time Complexity:

Best = O(n log n)

Average = O(n log n)

Worst = O(n log n)

‚úî Space Complexity:

O(n) (because merging uses extra temp array)

‚úî Merge Sort is:

2. How Merge Sort Works (Simple Explanation)
Step-1: Divide

Keep splitting array until each sub-array has one element.

Step-2: Combine/Merge

Take two sorted arrays ‚Üí compare 1st element of each ‚Üí build final sorted array.

                                   [8,3,1,7,0,10,2]
                                   /              \
                          [8,3,1]                     [7,0,10,2]
                         /      \                    /          \
                       [8]     [3,1]             [7,0]         [10,2]
                               /   \             /   \          /    \
                             [3]   [1]         [7]   [0]     [10]   [2]

                               ‚Üì merge           ‚Üì merge       ‚Üì merge
                             [1,3]              [0,7]         [2,10]

                       ‚Üì merge                                   ‚Üì merge
                     [1,3,8]                                    [0,2,7,10]

                                      ‚Üì FINAL MERGE
                          ---------------------------------
                          [0,1,2,3,7,8,10]


```python
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    mid=len(arr)//2
    left=merge_sort(arr[:mid])
    right=merge_sort(arr[mid:])
    return merge(left,right)
def merge(left,right):
    result=[]
    i=0
    j=0
    while i<len(left) and j<len(right):
        if left[i]<right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    while i<len(left):
        result.append(left[i])
        i+=1
    while j<len(right):
        result.append(right[j])
        j+=1
    return result
print(merge_sort([100,0,1,4]))
```
