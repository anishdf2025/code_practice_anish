# A decorator is a wrapper that modifies the behavior of a function.

```python
import logging
import time
from functools import wraps

logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(message)s")

def log_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()

        logging.info(f"Running '{func.__name__}'")
        logging.info(f"Args: {args}, Kwargs: {kwargs}")

        result = func(*args, **kwargs)

        end = time.time()
        logging.info(f"Result: {result}")
        logging.info(f"Execution time: {end - start:.5f} sec")

        return result
    return wrapper
@log_decorator
def multiply(x, y):
    """Multiply two numbers"""
    return x * y

multiply(4, 7)

```

```python
import time
def log(func):
    def anish():
        print(f"Calling {func.__name__}")
        start=time.time()
        func()
        end=time.time()
        total=end-start
        print(total)
    return anish

@log
def greet():
    time.sleep(2)
    print("Hi Anish!")

@log
def test():
    for i in range(1000000):
        pass

test()
greet()
```

```python
import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print("Execution time:", end - start)
    return wrapper

@timer
def test():
    for i in range(1000000):
        pass

test()
```

```python
def repeat(n):
    def decorator(func):
        def wrapper():
            for _ in range(n):
                func()
        return wrapper
    return decorator

@repeat(3)
def hello():
    print("Hello")

hello()
```
