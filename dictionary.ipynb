1. What is a Dictionary in Python? (Definition)

A dictionary in Python is an unordered, mutable, dynamic, key-value data structure.

‚úî Key Features

Stores data as: { key: value }

Keys must be unique

Keys must be hashable (immutable: str, int, float, tuple, bool)

Values can be anything (list, dict, int, obj, function)

Fast lookup: O(1) average time (hash table)

```python
d = dict(name="Anish", age=25)
print(d["name"])
print(d["age"])
```

```python
d={"name":"Anish","age":25}
print(d["age"])
print(d["name"])
```

```python
d = {
    "name": "Anish",
    10: "integer key",
    5.5: "float key",
    True: "boolean key",
    (1, 2, 3): "tuple key"
}
print("Original:", d)
```

```python
print(d[True])
print(d[5.5])
print(d[10])
```

```python
for v in d.values():
    print(v)
```

```python
print(d[(1,2,3)])
print(10)
print("dictionary:",d.values())
```

```python
print(list(d.keys()))
```

```python
print(list(d.values()))
```

```python
d={
    "Name":"Anish",
    "Age":28,
    10:"integer",
    True:"boolean value",
    5.5:"float key value",
    (1,2,3):"tuple"
}


print("All values of dictionary")
print(f"{d.values()} \n")
print("All key value")
print(f"{d.keys()} \n")
print("print the value via loop")
for v in d.values():
    print(f"{v}")
```

The dictionary method pop() in Python is used to:

üü¶ REMOVE a key‚Äìvalue pair from the dictionary AND return its value.
‚úÖ POP ‚Üí Remove + Return Value

Syntax:

dictionary.pop(key)


What it does:

Looks for the key

Removes that key‚Äìvalue pair from dictionary

Returns the value of that key

If key does NOT exist ‚Üí KeyError

```python
d.pop('Age')

```

Example 1: Normal pop

```python
dic1={"Name":"Anish","Age":28}
```

Example 1: Normal pop

```python
value=dic1.pop('Name')
print(value)
```

Example 2: Pop with default value (NO ERROR)

```python
value=dic1.pop('Name',"Not found")
print(value)
```

| Feature                       | Explanation                             |
| ----------------------------- | --------------------------------------- |
| **Removes and returns value** | Unlike `del`, `pop` gives you the value |
| **Optional default**          | Prevents KeyError                       |
| **Useful in algorithms**      | Extract-and-remove pattern              |

| Method         | Use Case                                          |
| -------------- | ------------------------------------------------- |
| **pop(key)**   | Remove a **specific** key and get its value       |
| **popitem()**  | Remove the **last inserted** key-value pair       |
| **clear()**    | Empty the entire dictionary                       |
| **update()**   | Add or modify multiple keys at once / merge dicts |
| **del d[key]** | Remove a specific key (does NOT return value)     |

‚úÖ 2Ô∏è‚É£ Update Dictionary Value Based on Condition
Question

Increase marks by 5 only for marks < 50.

```python
students = {"ram": 85, "shyam": 40, "rita": 92, "amit": 33}
for k,v in students.items():
    if v<50:
        students[k]=v+5
print(students)
```

‚úÖ 3Ô∏è‚É£ Reverse Lookup (Find Key Using Value)
Question

```python
d = {"a": 1, "b": 2, "c": 3,"d":2}
# find key for 2
for k,v in d.items():
    if v==2:
        print(k)
```
