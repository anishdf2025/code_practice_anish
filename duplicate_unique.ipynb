# Find pair of numbers whose sum is K.

```python
def sum_k(arr, K):
    for i in range(len(arr)-2):
        for j in range(i+1, len(arr)-1):
            for k in range(j+1, len(arr)):
                if arr[i] + arr[j] + arr[k] == K:
                    return arr[i], arr[j], arr[k], i, j, k
    return None

list_2 = [1, 0, 1, 4, 5, 6, 7]

try:
    a, b, c, i, j, k = sum_k(list_2, 1)
    print(a, b, c, i, j, k)
except TypeError:
    print("No triplet found that sums to given K")
```

# Given a list, shift all zeros to the end.

```python
def zero_shift(arr):
    non_zero=[]
    zeros=[]
    
    for num in arr:
        if num==0:
            zeros.append(num)
        else:
            non_zero.append(num)
    return zeros+non_zero
arr=[1,0,2,3,0,-1,-2,-3]
result=zero_shift(arr)
print(result)
            
```
