```python
def vowel_frequency(input):
    result=""
    count=0
    for ch in input:
        if ch=='A' or ch=='a' or ch=='e' or ch=='E' or ch=='i' or ch=='I' or ch=='O' or ch=='o' or ch=='U' or ch=='u':
            count+=1 
            result+=ch
    return count,result
input="ANISH"
print(vowel_frequency(input))
```
