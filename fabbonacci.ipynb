# Factorial of number

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
result=factorial(0)
print(result)
```

```python
def factorial_for_loop(n):
    result=1
    if n==0 or n==1:
        return 1
    else:
        for i in range(1,n+1):
            result*=i
        return result
result=factorial_for_loop(5)
print(result)

```

# Write a program that will reverse a four digit number.Also it checks whether the reverse is true.

```python
def sqrt(n):
    if n<0:
        return ValueError("number is not valid")
    low=0
    high=n
    precision=0.00000000001
    while (high-low)>precision:
        mid=(high+low)/2
        if mid*mid>n:
            high=mid
        else:
            low=mid
    return (high+low)/2
user_20=float(input("Enter the number"))
result_20=sqrt(user_20)
print(result_20)
```
