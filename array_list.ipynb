Problem 3: Remove all even numbers from an array<br>
‚úî Definition

You must delete every number that is divisible by 2.

‚úî Solution

```python
from array import array
a3=array('i',[24,28,301,2,1])
for x in a3:
    if x%2==0:
        a3.remove(x)
print(a3)
```

‚úÖ Problem 4: Reverse an array without using reverse()<br>
‚úî Definition

We manually reverse array using a loop.

‚úî Solution

```python
from array import array
a5=array('i',[9,10,22,45,32])
res=array('i')
for x in range(len(a5)-1,-1,-1):
    res.append(a5[x])
print(res)
```

‚úÖ Problem 5: Find maximum & minimum in array<br>
‚úî Definition

Traverse the array and find:

Largest number

Smallest number

```python
from array import array
a6=array('i',[90,102,1,0,22])
min=a6[0]
max=a6[0]
for x in a6:
    if x>max:
        max=x
    elif x<min:
        min=x
print(min,max)
```
