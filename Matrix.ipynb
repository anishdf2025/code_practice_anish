# Matrix Operation

```python
Matrix=[
    [1,2,3],
    [4,6,8],
    [11,12,23]
]
row_length=len(Matrix)
column_length=len(Matrix[0])
print(row_length)
print(column_length)
```

# Sum of principal digonal elements

```python
result=0
for i in range(row_length):
    result+=Matrix[i][i]
print(result)
```

# sum of two matrix 

```python
m1=[
    [1,2,3,4],
    [2,4,6,7],
    [9,1,0,3]
    ]
m2=[
    [31,43,56,1],
    [2,7,9,2],
    [9,2,4,5]
    ]
sum=[]
for i in range(len(m1)):
    res=[]
    for j in range(len(m1[0])):
        res.append(m1[i][j]+m2[i][j])
        
    sum.append(res)
for row in sum:
    print(row)
        
```

# Substraction of two matrix

```python
m3=[[1,2,3],
    [3,1,4],
    [2,4,7]
    ]
m4=[[1,2,3],
    [3,1,6],
    [2,4,7]
    ]


sub=[]
for i in range(len(m3)):
    res_sub=[]
    for j in range(len(m3[0])):
        res_sub.append(m3[i][j]-m4[i][j])
    
    sub.append(res_sub)  
for row in sub:
    print(row)  
```

# Transpose of Matrix

```python
m5=[
    [1,2,4],
    [2,5,7],
    [9,5,3]
    ]
result=[]
for i in range(len(m5[0])):
    transpose=[]
    for j in range(len(m5)):
        transpose.append(m5[j][i])
    result.append(transpose)
for row in result:
    print(row)
    
```

```python
m5=[[1,2,3],
    [3,1,4],
    [2,4,7]
    ]
m6=[[1,2,3],
    [3,1,6],
    [2,4,7]
    ]
for i in range(len(m5)):
    res=[]
    for j in range(len(m5[0])):
        sum=0
        for k in range(len(m5)):
            sum+=m5[i][k]*m6[k][j]
        res.append(sum)
    print(res)
            
```
