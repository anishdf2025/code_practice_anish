## Python Basic practice

*args & **kwargs 

Addition of multiple number

```python
def add_num(*args):
    sum=0
    for num in args:
        sum+=num
    return sum
print(add_num(1,2,3,4))
```

Multiplication of number 

```python
def mul_num(*args):
    mul=1
    for num in args:
        mul*=num
    return mul
print(mul_num(2,3,4)) 
```

**kwargs example

```python
def print_patient_information(**kwargs):
    return kwargs
print(print_patient_information(Name="ANISH",Age=29,country="INDIA"))
```

*args and **kwargs example

```python
def pizza_order(size, *lopping, **details):
    print(f"Size of pizza: {size}-inches")
    print("Loppings:", lopping)
    print("Details:", details)

pizza_order(12, "cheese", "tomato", "pepper", crust="thin", city="Bengaluru")

```

```python
def pizza_order(size,*lopping,**details):
    return {
        "size":size,
        "lopping":lopping,
        "details":details
    }
pizza=pizza_order(12,"cheese","butter",Crust="thin",extra_item=True)
print(f"order of pizza {pizza}")
```
