1️⃣ Bubble Sort Theory

Bubble Sort is a simple comparison-based sorting algorithm.<br>

It works by repeatedly swapping adjacent elements if they are in the wrong order.<br>

The largest elements “bubble” to the end in each pass, hence the name.<br>

Sorting continues until no swaps are needed.<br>

Steps (Ascending Order):

Start at the first element.

Compare the current element with the next element.

Swap if the current element is greater than the next.

Move to the next element and repeat until the end of the array.

Repeat passes for the remaining unsorted portion of the array until sorted.

Example (Array: [5, 1, 4, 2, 8])

Pass 1:

Compare 5 and 1 → swap → [1, 5, 4, 2, 8]<br>
Compare 5 and 4 → swap → [1, 4, 5, 2, 8]<br>
Compare 5 and 2 → swap → [1, 4, 2, 5, 8]<br>
Compare 5 and 8 → no swap → [1, 4, 2, 5, 8]<br>


Pass 2:

Compare 1 and 4 → no swap <br>
Compare 4 and 2 → swap → [1, 2, 4, 5, 8]<br>
Compare 4 and 5 → no swap<br>
Compare 5 and 8 → no swap<br>


Pass 3:

Compare 1 and 2 → no swap<br>
Compare 2 and 4 → no swap<br>
Compare 4 and 5 → no swap<br>
Compare 5 and 8 → no swap<br>


✅ Array is now sorted: [1, 2, 4, 5, 8]

```python
# Increasing order of list
def bubble_sort_icr(arr):
    n=len(arr)
    for i in range(n):
        for j in range(n-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
arr1=[90,2,5,12,32]
res=bubble_sort_icr(arr1)
print(res)
```

```python
# Decreasing order of list
def bubble_sort_dcr(arr):
    n=len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]<arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    return arr
arr2=[90,2,5,12,32]
res1=bubble_sort_dcr(arr2)
print(res1)


```

1️⃣ Loops Analysis

Outer loop: for i in range(n) → runs n times

Inner loop: for j in range(n-1) → runs n-1 times each outer loop iteration

Total comparisons = n * (n-1) ≈ n²

```python
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        swap=False
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                swap=True
        if not swap:
            break
    return arr
arr3=[90,2,5,12,32]  
res3=bubble_sort(arr3)
print(res3)
```
