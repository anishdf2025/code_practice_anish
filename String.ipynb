4. How to reverse a string without using slicing or reverse()?<br>
Interviewer expects logic, not built-in shortcuts.

```python
P="ANISH"
rev=""
for x in P:
    rev=x+rev
print(rev)
```

Palindrome

```python
s="ANA"
i=0
j=len(s)-1
is_pal=True
while i<j:
    if s[i]!=s[j]:
        is_pal=False
        break
    i+=1
    j-=1
print(is_pal)
```
